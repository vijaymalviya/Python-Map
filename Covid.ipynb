```python
df = pd.read_csv("cov.csv")
```

```python
m = folium.Map(
    zoom_start=1.5,width=1000,height=750,location=[20.5937,78.9629],
    tiles='Stamen Terrain'
)
for row in df.itertuples():
    lat=getattr(row, "Latitude")
    long=getattr(row, "Longitude")
    Confirm= int(row[4])
    name = row[1]
  # name=getattr(row, "Name")
  # tooltip = f"{Name} - {Confirmed}"
    radius = 30 if Confirm/10>30 else Confirm/10

    if Confirm>0:
        folium.Marker(
        location=[lat,long],
   #    popup=Confirm,
         popup = [Confirm,name],
         icon=folium.Icon(color='orange', icon='info-sign')
         ).add_to(m)
```
